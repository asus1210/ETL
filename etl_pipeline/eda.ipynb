```python
def add_season(org_df : pd.DataFrame) -> pd.DataFrame:
    org_df["Date"] = pd.to_datetime(org_df["Date"], format="mixed" ,dayfirst=True)
    new_df = org_df.copy()
    season_year = "-".join(str(year) for year in new_df["Date"].dt.year.unique())
    new_df.loc[:,"SeasonYear"] = season_year
    return new_df    
```

```python
modified_dataframe_list = list()
for org_df in dataframe_list:
    modified_dataframe_list.append(add_season(org_df=org_df))
whole_df = pd.concat(modified_dataframe_list, sort=False)
```

```python
def update_points(row):
    home_team = row["HomeTeam"]
    away_team = row["AwayTeam"]
    goal_scored_home = row["FTHG"]
    goal_conceded_home = row["FTAG"]
    goal_scored_away = row["FTAG"]  # Away team's goals
    goal_conceded_away = row["FTHG"]  # Away team's goals conceded
    ftr = row["FTR"]
    season_year = row["SeasonYear"]
    
    table.loc[(table["Team"] == home_team) & (table["SeasonYear"] == season_year), "GF"] += goal_scored_home
    table.loc[(table["Team"] == home_team) & (table["SeasonYear"] == season_year), "GA"] += goal_conceded_home
    table.loc[(table["Team"] == home_team) & (table["SeasonYear"] == season_year), "GD"] += (goal_scored_home - goal_conceded_home)

    # Update away team stats
    table.loc[(table["Team"] == away_team) & (table["SeasonYear"] == season_year), "GF"] += goal_scored_away
    table.loc[(table["Team"] == away_team) & (table["SeasonYear"] == season_year), "GA"] += goal_conceded_away
    table.loc[(table["Team"] == away_team) & (table["SeasonYear"] == season_year), "GD"] += (goal_scored_away - goal_conceded_away)
    
    if ftr == "H":
        table.loc[(table["Team"] == home_team) & (table["SeasonYear"] == season_year), "Points"] += 3
        table.loc[(table["Team"] == home_team) & (table["SeasonYear"] == season_year), "W"] += 1
        table.loc[(table["Team"] == away_team) & (table["SeasonYear"] == season_year), "L"] += 1
    elif ftr == "A":
        table.loc[(table["Team"] == away_team ) & (table["SeasonYear"] == season_year), "Points"] += 3
        table.loc[(table["Team"] == away_team) & (table["SeasonYear"] == season_year), "W"] += 1
        table.loc[(table["Team"] == home_team) & (table["SeasonYear"] == season_year), "L"] += 1
    elif ftr == "D":
        table.loc[(table["Team"].isin([home_team, away_team])) & (table["SeasonYear"] == season_year) , "Points"] += 1
        table.loc[(table["Team"].isin([home_team, away_team])) & (table["SeasonYear"] == season_year) , "D"] += 1

whole_df[
    ["SeasonYear", "HomeTeam", "AwayTeam", "FTHG", "FTAG","FTR"]
    ].apply(update_points, axis=1)
```

```python
def update_points(row):
    home_team = row["HomeTeam"]
    away_team = row["AwayTeam"]
    goal_scored_home = row["FTHG"]
    goal_conceded_home = row["FTAG"]
    goal_scored_away = row["FTAG"]  # Away team's goals
    goal_conceded_away = row["FTHG"]  # Away team's goals conceded
    ftr = row["FTR"]
    season_year = row["SeasonYear"]

    # Update home team stats
    home_team_mask = (table["Team"] == home_team) & (table["SeasonYear"] == season_year)
    table.loc[home_team_mask, "GF"] += goal_scored_home
    table.loc[home_team_mask, "GA"] += goal_conceded_home
    table.loc[home_team_mask, "GD"] += (goal_scored_home - goal_conceded_home)

    # Update away team stats
    away_team_mask = (table["Team"] == away_team) & (table["SeasonYear"] == season_year)
    table.loc[away_team_mask, "GF"] += goal_scored_away
    table.loc[away_team_mask, "GA"] += goal_conceded_away
    table.loc[away_team_mask, "GD"] += (goal_scored_away - goal_conceded_away)

    # Update points and other match-related statistics
    if ftr == "H":
        table.loc[home_team_mask, "Points"] += 3
        table.loc[home_team_mask, "W"] += 1
        table.loc[away_team_mask, "L"] += 1
    elif ftr == "A":
        table.loc[away_team_mask, "Points"] += 3
        table.loc[away_team_mask, "W"] += 1
        table.loc[home_team_mask, "L"] += 1
    elif ftr == "D":
        draw_mask = table["Team"].isin([home_team, away_team]) & (table["SeasonYear"] == season_year)
        table.loc[draw_mask, "Points"] += 1
        table.loc[draw_mask, "D"] += 1

# Create an empty points table DataFrame with appropriate columns before running the loop
# Example:
# table = pd.DataFrame(columns=["Team", "SeasonYear", "GF", "GA", "GD", "Points", "W", "L", "D"])

# Loop through match data to update the points table
for _, row in whole_df.iterrows():
    update_points(row)
```
